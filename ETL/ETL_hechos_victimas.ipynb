El dataframe de hechos tiene 696 entradas (filas) y 21 columnas.

El dataframe de victimas tiene 717 entradas (filas) y 10 columnas.

Observamos que se pueden unir ambos dataframes mediante el ID del hecho, pero antes de eso , vamos a eliminar las columnas del dataframe de victimas que ya esten presentes en el dataframe de hechos.

```python
victimas_columnas_repetidas = ['ID_hecho', 'ROL', 'SEXO','EDAD']

# Eliminar las columnas especificadas
df_victimas = df_victimas[victimas_columnas_repetidas]
```

Ya que hay hechos que involucran mas de una victima, verifiquemos la calidad del dataset en general antes de hacer el join. De modo que la suma de victimas de la tabla de hechos, debe ser igual a la cantidad de registros del dataframe de victimas.

```python
total_victimas_en_hechos = df_hechos['N_VICTIMAS'].sum()
total_registros_en_victimas = len(df_victimas)

if total_victimas_en_hechos == total_registros_en_victimas:
    print("El número total de víctimas coincide correctamente.")
else:
    print("Hay una discrepancia entre el número total de víctimas en df_hechos y el número de registros en df_victimas.")
```

```python
# Unir dataframes mediante el ID del hecho
df = pd.merge(df_hechos, df_victimas, left_on='ID', right_on='ID_hecho', how='left')

df = df.drop("ID_hecho", axis=1)
```

```python
df.info()
```

La primera intervencion sera estandarizar la escritura de todos los valores del set de datos a minúsculas para obtener consistencia, facilidad en busquedas y comparaciones. Los encabezados de las columnas se modificaran para que esten escritos completamente en mayusculas.

```python
# Encabezados en mayúsculas
df.columns = df.columns.str.upper()

# Seleccionar solo las columnas no numéricas
columnas_no_numericas = df.select_dtypes(exclude=['number']).columns

# Aplicar la conversión a minúsculas en esas columnas
df[columnas_no_numericas] = df[columnas_no_numericas].map(lambda x: x.lower() if isinstance(x, str) else x)
```

Realizaremos el proceso de limpieza teniendo en cuenta las situaciones más comunes:

Al final de este proceso de limpieza deberíamos tener un set de datos **íntegro**, listo para la fase de Análisis Exploratorio.

Agregaremos una columna extra llamada "DIA_SEMANA", calculada a partir de la fecha, que nos permitirá elaborar graficos con mayor facilidad mas adelante.

```python
from datetime import datetime

# Crear una nueva columna para el día de la semana
df["DIA_SEMANA"] = None

# Recorrer las filas del DataFrame
for index, row in df.iterrows():
    fecha_str = str(row["FECHA"])
    fecha_str_recortada = fecha_str.split()[0] 
    fecha_dt = datetime.strptime(fecha_str_recortada, "%Y-%m-%d")

    dia_num = fecha_dt.weekday()
    dias_semana = ["lunes", "martes", "miercoles", "jueves", "viernes", "sabado", "domingo"]
    dia_nombre = dias_semana[dia_num]
    df.loc[index, "DIA_SEMANA"] = dia_nombre

```

Cambiaremos el contenido de la columna "MM" para que tenga el nombre del mes en lugar del numero de mes, con el fin de facilitar la lectura en los gráficos.

```python
# Diccionario con los nombres de los meses en minúsculas
meses_minusculas = {1: "enero", 2: "febrero", 3: "marzo", 4: "abril", 5: "mayo", 6: "junio",
                    7: "julio", 8: "agosto", 9: "septiembre", 10: "octubre", 11: "noviembre", 12: "diciembre"}

# Reemplazar valores en la columna "MM" por el nombre del mes correspondiente
df["MM"] = df["MM"].replace(meses_minusculas)
```

Renombramos las columnas de año, mes y día para facilitar su entendimiento.

```python
df.rename(columns={'AAAA': 'AÑO', 'MM': 'MES', 'DD': 'DIA'}, inplace=True)
```

Con el fin de evitar redundancia en los datos, eliminaremos HH, ya que usamos "HORA" que incluye horas, minutos y segundos. Dejaremos la columna "FECHA", a pesar de tener las componentes en columnas separadas, para hacer uso de ella donde sea conveniente.

```python
columna_a_eliminar = ["HH"]

df = df.drop(columna_a_eliminar, axis=1)
```

Nos aseguramos de que la columna FECHA tenga un formato date y la columna HORA un formato time.

```python
df['FECHA'] = pd.to_datetime(df['FECHA'].dt.strftime('%Y-%m-%d'))

# Convertir la columna 'hora' a tipo string para manipularla
df['HORA'] = df['HORA'].astype(str)

# Identificar y extraer solo la hora
df['HORA'] = df['HORA'].apply(lambda x: x.split()[-1])
df['HORA'] = pd.to_datetime(df['HORA'], format='%H:%M:%S', errors='coerce').dt.time
```

Crearemos la variable categórica `rango_etario` a partir de la variable `EDAD` con rangos etarios comprendidos entre `0-17`, `18-29`, `30-44`, `45-64` y `65+`, con el objetivo de facilitar el análisis exploratorio de datos. Para esto, procedemos de la siguiente forma:

```python
# Convertimos la columna 'EDAD' a tipo de dato numérico
df['EDAD'] = pd.to_numeric(df['EDAD'], errors='coerce')

# Definimos los límites de los intervalos de edad
bins = [0, 18, 30, 45, 65, 100]

# Definimos las etiquetas de los intervalos de edad
labels = ['0-17', '18-29', '30-44', '45-64', '65+']

# Creamos una nueva columna 'rango_edad' con los intervalos de edad
df['RANGO_ETARIO'] = pd.cut(df['EDAD'], bins=bins, labels=labels, right=False)

# Agregamos la categoría 'SD' a la columna 'rango_edad'
df['RANGO_ETARIO'] = df['RANGO_ETARIO'].cat.add_categories('sin datos')
df[['EDAD','RANGO_ETARIO']] = df[['EDAD','RANGO_ETARIO']].fillna('sin datos')

# Verificamos el resultado
df[['EDAD','RANGO_ETARIO']].sample(5)
```
